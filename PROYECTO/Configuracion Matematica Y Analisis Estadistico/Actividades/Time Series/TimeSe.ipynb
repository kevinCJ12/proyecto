```python
import pandas as pd
import numpy as np
df=pd.read_csv('5AbrilTotalCases.csv')
df=df.fillna(method='ffill')
df['date']=pd.to_datetime(df['date'])
df=df.set_index('date')
```

```python
df=df.filter(items=['Mexico', 'Spain','Italy','Chile','United States','Switzerland','Argentina','Colombia'])
pd.set_option('display.max_columns', None) 
pd.set_option('display.expand_frame_repr', False)
```

```python
resumen2 = pd.DataFrame()
resumen2 = df.resample('M').last()
print(resumen2)
```

```python
resumen['S']=df.Mexico.resample('W-MON').last()
print(resumen['S'])
```

```python
resumen = pd.DataFrame()
resumen['PS']=df.Mexico.resample('W-MON').mean().round(2)
print(resumen)
```

```python
resumen['PSm']=df.Mexico.resample('W-MON').mean().min()
print(resumen)
```

```python
diferencias=df.diff()
diferencias=diferencias.abs()
print(list(diferencias))
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
cols_plot = ['Mexico', 'Spain','Italy','Chile','United States','Switzerland','Argentina','Colombia']
fig = make_subplots(rows=len(cols_plot), cols=1, start_cell="bottom-left")
i=1
j=0
for col in cols_plot:
 trace=go.Scatter(x=diferencias.index, y=diferencias[col],mode='lines',name=cols_plot[j])
 fig.append_trace(trace,i,1)
 i=i+1
 j=j+1
fig.show()
```
